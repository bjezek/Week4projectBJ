```python
# import Libaries 

import pandas as pd
import numpy as np
from pathlib import Path

%matplotlib inline
```

```python
# read csvfile

risk_return = pd.read_csv(Path("./whale_navs.csv"), index_col = "date", parse_dates = True, infer_datetime_format = True)
```

```python
# return first 5 of funds 
risk_return.head()
```

```python
#Calculate daily returns 

daily_returns = risk_return.pct_change().dropna()
daily_returns
```

```python
# return and plot s&p 500 daily returns 

daily_returns['S&P 500'].plot(figsize=(10,7), title = "S&P 500 daily returns")
```

```python
# caluculate soros capital cumulative returns

soros_cumulative_returns = (1+daily_returns["SOROS FUND MANAGEMENT LLC"]).cumprod()
soros_cumulative_returns
```

```python
#Calculate Paulson cumulative returns 

paul_cumulative_returns = (1+daily_returns["PAULSON & CO.INC."]).cumprod()
paul_cumulative_returns
```

```python
#Calculate Tiger Cumulative fund

tiger_cumulative_returns =(1+daily_returns["TIGER GLOBAL MANAGEMENT LLC"]).cumprod()
tiger_cumulative_returns
```

```python
#Calculate Berk cumulative returns

berk_cumulative_returns = (1 + daily_returns["BERKSHIRE HATHAWAY INC"]).cumprod()
berk_cumulative_returns
```

```python
#Calculate s&p500 Cumulative returns

sp500_cumulative_returns = (1+ daily_returns["S&P 500"]).cumprod()
sp500_cumulative_returns
```

```python
# days in trading 

days_in_trading = 252
```

```python
annual_portfolio_returns = daily_returns.mean() * np.sqrt(days_in_trading)
annual_portfolio_returns.sort_values()
```

```python
# calculate annual std of portfolios

annual_std_portfolio = daily_returns.std() * np.sqrt(days_in_trading)
annual_std_portfolio.sort_values()
```

```python
# Calculate portfolio returns over 60 day period, 2 months 

annual_portfolio_returns = daily_returns.rolling(window=60).mean()
annual_portfolio_returns.dropna()
```

```python
#plot sharp ratio of funds bar plot 

sharp_ratio_portfolio = annual_portfolio_returns / annual_std_portfolio
sharp_ratio_portfolio.plot.bar(figsize=(15,20), title = "Sharp Ratio for Soro, Tiger, Paul, Berk & S&P 500")
```

```python
#Plot rolling 60 day for sp500 

market_rolling_60_variance = daily_returns['S&P 500'].rolling(window=60).var()
market_rolling_60_variance.plot(figsize=(10,7), title= " S&P 500 Rolling 60 Variance")
market_rolling_60_variance.head().dropna()
```

```python
#Plot soros 60 rolling day cov

soros_rolling_60_covariance = daily_returns['SOROS FUND MANAGEMENT LLC'].rolling(window=60).cov(daily_returns['S&P 500'])
soros_rolling_60_covariance.plot(figsize=(10,7), title= " Soros Rolling 60 Covariance")
```

```python
#Plot berk 60 day rolling cov

berk_rolling_60_covariance = daily_returns['BERKSHIRE HATHAWAY INC'].rolling(window=60).cov(daily_returns['S&P 500'])
berk_rolling_60_covariance.plot(figsize=(10,7), title= " Berk Rolling 60 Covariance")
```

```python
#Plot tiger rolling 60 day cov

tiger_rolling_60_covariance = daily_returns['TIGER GLOBAL MANAGEMENT LLC'].rolling(window=60).cov(daily_returns['S&P 500'])
tiger_rolling_60_covariance.plot(figsize=(10,7), title= " Tiger Rolling 60 Covariance")
```

```python
#Plot paul 60 day rolling cov

paul_rolling_60_covariance = daily_returns['PAULSON & CO.INC.'].rolling(window=60).cov(daily_returns['S&P 500'])
paul_rolling_60_covariance.plot(figsize=(10,7), title= " Paul Rolling 60 Covariance")
```
